### CRATING THE COMPLEX VALUES FROM THE REAL AND IMAGINARY PORTIONS OF EACH DATASET

### THEN CONVERTING TO DB USING 20*LOG10(ABS(POWER/memory))~("SAME USED FOR REFLECTIONS")

### BUILDING THE FUNCTIONS TO CALCULATE THE DB MAGNITUDES FORM THE AVAILABLE DATA WITH MEMORY 

```python
for i in range(len(one_inch_microstrip_data)):
    def db_magnitude_memory_S21(data):
        power                              = (np.asarray(data['real_S21']) + 1j * 
                                             np.asarray(data['imaginary_S21']))
        
        memory                             = (np.asarray(data['real_memory_S21']) + 1j * 
                                             np.asarray(data['imaginary_memory_S21']))
        
        magnitude_memory_S21                          = 20*log10(abs(power/memory))
        return magnitude_memory_S21
    
for i in range(len(one_inch_microstrip_data)):
    def db_magnitude_memory_S12(data):
        power                              = (np.asarray(data['real_S12']) + 1j * 
                                             np.asarray(data['imaginary_S12']))
        
        memory                             = (np.asarray(data['real_memory_S12']) + 1j * 
                                             np.asarray(data['imaginary_memory_S12']))
        
        magnitude_memory_S12                          = 20*log10(abs(power/memory))
        return magnitude_memory_S12 

for i in range(len(one_inch_microstrip_data)):
    def db_magnitude_memory_S11(data):
        power                              = (np.asarray(data['real_S11']) + 1j * 
                                             np.asarray(data['imaginary_S11']))
        
        memory                             = (np.asarray(data['real_S11']) + 1j * 
                                             np.asarray(data['imaginary_S11']))
        
        magnitude_memory_S11                          = 20*log10(abs(power/memory))
        return magnitude_memory_S11

for i in range(len(one_inch_microstrip_data)):
    def db_magnitude_memory_S22(data):
        power                              = (np.asarray(data['real_S22']) + 1j * 
                                             np.asarray(data['imaginary_S22']))
        
        memory                             = (np.asarray(data['real_S22']) + 1j * 
                                             np.asarray(data['imaginary_S22']))
        
        magnitude_memory_S22                          = 20*log10(abs(power/memory))
        return magnitude_memory_S22
```

```python
for i in range(len(one_inch_microstrip_data)):
    def db_magnitude_S21(data):
        magnitude_S21                              = 20*log10(abs(np.asarray(data['real_S21']) + 1j * 
                                             np.asarray(data['imaginary_S21'])))

        return magnitude_S21
    
for i in range(len(one_inch_microstrip_data)):
    def db_magnitude_S12(data):
        magnitude_S12                          = 20*log10(abs(np.asarray(data['real_S12']) + 1j * 
                                             np.asarray(data['imaginary_S12'])))
        return magnitude_S12 

for i in range(len(one_inch_microstrip_data)):
    def db_magnitude_S11(data):
        magnitude_S11                          = 20*log10(abs(np.asarray(data['real_S11']) + 1j * 
                                             np.asarray(data['imaginary_S11'])))
        return magnitude_S11

for i in range(len(one_inch_microstrip_data)):
    def db_magnitude_S22(data):
        magnitude_S22                              = 20*log10(abs(np.asarray(data['real_S22']) + 1j * 
                                             np.asarray(data['imaginary_S22'])))
        return magnitude_S22
```

```python
c = 3.99e8
lam  = (c/padins_one_inch_gcpw_data['frequency'])

def martix_exp(S_o, lam):
    theta = (2*np.pi*S_o)/(lam)
    return theta
def complex_matrix_exp(S_o, lam):
    complex_theta = (((2*np.pi*S_o)+1j)/(lam))
    return complex_theta

def vee_bar(frequency, t, phi, V_o):
    V = V_o*(cos(((2*np.pi*frequency)*t) + phi)+1j*(sin(((2*np.pi*frequency)*t) + phi)))
    return V

##transmission line matrix
transmission = np.array([[e**(complex_matrix_exp(1,100)), 0], [0, e**(complex_matrix_exp(1,100))]])
constant     = np.array([[vee_bar((padins_one_inch_gcpw_data['frequency']),7,45, 20), 0]])

solution     = (np.multiply(transmission,constant))

```
